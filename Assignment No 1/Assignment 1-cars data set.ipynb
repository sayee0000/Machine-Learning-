```python
fuel_type = ['CNG', 'Diesel', 'Petrol', 'LPG', 'Electric']
no_of_cars = [62, 3852, 3325,12, 2]
plt.bar(fuel_type, no_of_cars, color='grey')
plt.xlabel('Fuel Type')
plt.ylabel('Number of Cars')
plt.title('Count of Cars by Fuel Type')
plt.show()
```
