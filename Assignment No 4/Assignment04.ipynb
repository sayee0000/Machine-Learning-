```python
import pandas as pd
df=pd.read_csv(r"C:\Users\HP\Downloads\datalab_export_2025-08-12 13_51_24.csv")
df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(data=df,x='purpose',hue='not.fully.paid')
plt.xticks(rotation=45, ha='right');
```

```python
pre_df = pd.get_dummies(df,columns=['purpose'],drop_first=True)
pre_df.head()
```

```python
from sklearn.model_selection import train_test_split

X = pre_df.drop('not.fully.paid', axis=1)
y = pre_df['not.fully.paid']

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.33, random_state=125
)
```

```python
from sklearn.metrics import (
    accuracy_score,
    confusion_matrix,
    ConfusionMatrixDisplay,
    f1_score,
    classification_report,
)
# First, define your model (this is just an example - use your actual model)
from sklearn.ensemble import RandomForestClassifier  # or any other model you're using
model = RandomForestClassifier()
model.fit(X_train, y_train)  # Assuming X_train and y_train are defined

# Now predict using the trained model
y_pred = model.predict(X_test)

accuray = accuracy_score(y_test, y_pred)  # Note: target (y_test) should be first parameter
f1 = f1_score(y_test, y_pred, average="weighted")  # Note: target (y_test) should be first parameter

print("Accuracy:", accuray)
print("F1 Score:", f1)
```

```python
labels = ["Fully Paid", "Not fully Paid"]
cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=labels)
disp.plot();
```

```python
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    confusion_matrix,
    ConfusionMatrixDisplay,
    f1_score,
    classification_report,
)
from sklearn.ensemble import RandomForestClassifier

# Train the model
model = RandomForestClassifier()
model.fit(X_train, y_train)

# Predict
y_pred = model.predict(X_test)

precision = precision_score(y_test, y_pred, average="weighted", zero_division=0)
recall = recall_score(y_test, y_pred, average="weighted", zero_division=0)
print("Precision:", precision)
print("Recall:", recall)

```
